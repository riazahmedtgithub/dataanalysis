# Introduction about the data set

The dataset has been sourced from https://www.kaggle.com/datasets/rashikrahmanpritom/heart-attack-analysis-prediction-dataset

This dataset contains the below values and the probability of heart attack.  Columns present in the dataset are:

Age : Age of the patient  
Sex : Sex of the patient  
exang: Exercise induced angina (1 = yes; 0 = no)  
ca: Number of major vessels (0-3)  
cp : Chest Pain type chest pain type  

Value 1: typical angina  
Value 2: atypical angina  
Value 3: non-anginal pain  
Value 4: asymptomatic  
trtbps : resting blood pressure (in mm Hg)  

chol : Cholestoral in mg/dl fetched via BMI sensor  
fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)  
rest_ecg : Resting electrocardiographic results  

Value 0: normal  
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)  
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria  
thalach : maximum heart rate achieved  

target :   

Value 0= Less chance of heart attack   
Value 1= More chance of heart attack  

# Questions

Following are the couple of questions that we are trying to answer by plotting appropriate graphs.  

Is there a relation between resting blood pressure and cholestoral.  
Comparison of resting blood pressure and cholestoral between male and female.  
Distribution of Cholestoral level data.  
Who is more vulnerable of getting heart attack, Male or Female?  
Finding the central tendency, outliers and spread of age data.  

```python
# Load the dataset as a Pandas data frame.

import pandas as pd

heart_df = pd.read_csv(r"C:\Users\Riaz\Desktop\MSDS\Data Mining\Week 2\heart.csv")

# Required libraries are imported and using the read_csv function the dataframe has been loaded from csv file.
heart_df.head(5)
```

# Scatter plot

Scatter plot between resting blood pressure and cholestoral.  
We are trying to find out if there is any relation between resting blood pressure and cholestoral.  
As seen from the below plot, it is random.  There is no definite relationship between them.  

```python
import matplotlib.pyplot as plt
plt.scatter('trtbps','chol',data=heart_df)
plt.title("Scatter plot between resting blood pressure and cholestoral")
plt.xlabel("Resting blood pressure in mm Hg")
plt.ylabel("Cholestoral levels in mg/dl")
```

# Box plot

Plotting a box plot for Cholestoral levels to find out the distribution of data.  
From the below plot, it is found out that,  

The median level of Cholestoral is around 225 mg/dl.  
First quartile is 210 mg/dl and third quartile is around 375 mg/dl.  
There are few points which are lying out of whiskers (above 360mg/dl).  They are considered as outliers.  


```python
plt.boxplot('chol', data=heart_df, capprops=dict(color='brown'),flierprops=dict(marker='x',color='red'))
plt.title("Box plot for Cholestoral levels")
plt.ylabel("Cholestoral levels in mg/dl")
```

# Bar plot

Plotting a bar plot between Male and Female to find who is more vulnerable of getting heart attack.  
From the below graph it is clear from the data that men are more prone to get heart attack rather than females.  

```python
# I have used the groupby function and got the sum of the output column, which represents getting a heart attack or not.

plt.bar(heart_df.groupby('sex')['output'].sum().index,heart_df.groupby('sex')['output'].sum().values)
plt.xticks([])
plt.text(-0.1, 40, 'Female', fontsize=12, color='yellow')
plt.text(0.9, 40, 'Male', fontsize=12, color='yellow')
plt.title("Bar plot between gender showing who are more prone to heart attack")
plt.xlabel("Gender")
plt.ylabel("Number of observations found")
```

# Histogram

I have plotted the histogram of the age column and following has been found out.  
It follows a normal and unimodal distribution.  
Mode is around 60 years.  
Mean is around 55 years.  
Spread is from 30 to 75 years.  
Most of the collected age values lie from 45 to 65 years.  
No outliers are to be found.  


```python
plt.hist(heart_df['age'],bins='auto')
plt.title("Histogram of age")
plt.xlabel("Age in years")
plt.ylabel("Frequency")
```

# Violin plot of age

Violin plot has also been plotted and it also corresponds to the above observation found from histogram  

```python
import seaborn as sns
sns.violinplot(data=heart_df['age'])
plt.title("Violin plot of age")
plt.ylabel("Age in years")
```

# Bubble plot

Plotting a bubble plot with legend information to find out how Male/Female groups compare with cholestoral and resting blood pressure levels.  
From the below plot, it is seen that very marginally Clolestoral levels are elevated in men than women.   

```python
scatter=plt.scatter('trtbps','chol',c='sex',cmap='cool',alpha=0.75,edgecolors='w',data=heart_df)
handles = [
    plt.scatter([], [], edgecolor='w', facecolor='cyan', label='Male'),plt.scatter([], [], edgecolor='w', facecolor='magenta', label='Female')]
plt.legend(handles=handles)
plt.title("Scatter plot between resting blood pressure and cholestoral")
plt.xlabel("Resting blood pressure in mm Hg")
plt.ylabel("Cholestoral levels in mg/dl")
```
